# RL-Based Active Learning for Medical Image Annotation
---

### Project Context

The project trains a DQN-based reinforcement learning agent to decide which unlabeled
medical images a human expert should annotate, under a strict labeling budget. 

### Dataset

| Property | Value |
|----------|-------|
| Source | MedMNIST v2 (Yang et al., 2023) |
| Primary | PneumoniaMNIST (binary: Normal vs Pneumonia) |
| Secondary | BreastMNIST (binary: Malignant vs Benign/Normal) |
| Image size | 28 x 28 pixels, grayscale |
| Access | Public, no credentialing required — auto-downloaded via `medmnist` pip package |
| License | CC BY 4.0 |
| Paper | https://doi.org/10.1038/s41597-022-01721-8 |

---
This notebook implements a **state-of-the-art RL-based active learning** framework for medical image annotation, combining:
- **Active Learning (AL)**: Strategically selecting which unlabeled samples to annotate
- **Reinforcement Learning (RL)**: A DQN agent that learns annotation decisions as an MDP
- **SOTA Improvements**: 8-dim budget-aware state, shaped rewards, soft target updates, Prioritised Experience Replay, and Dueling Double DQN

### Research Context
> *Fang et al. [2017] showed RL policies can outperform entropy baselines on classification tasks.*  
> *Contardo et al. modelled sample selection as an MDP — RL can allocate annotation effort more strategically than fixed heuristics.*  
> *This notebook extends those findings to **medical imaging** under strict budgets.*

### MDP Formulation
| Element | Value |
|---------|-------|
| State | 8-dim: confidence, entropy, margin, budget ratio, AUC, labeled ratio, step progress, class imbalance |
| Action | 0 = predict (free) · 1 = request annotation (costs budget) |
| Reward | Shaped: entropy proxy + 10× ΔAUC bonus − label penalty |
| Episode | One full annotation budget cycle |

### Strategies Compared
| Strategy | Type |
|----------|------|
| Random Sampling | Baseline |
| Margin Sampling | Heuristic |
| Entropy Sampling | Heuristic (myopic uncertainty) |
| BALD (MC-Dropout) | Bayesian uncertainty |
| DQN | RL — budget-aware learned policy |
| Double DQN | RL — reduced overestimation bias |
| Dueling DQN | RL — decoupled value/advantage |

## 2. Dataset Pipeline

### Dataset Choice

**Primary — PneumoniaMNIST**
Derived from the Guangzhou Women and Children's Medical Center chest X-ray dataset
(Kermany et al., 2018). Binary task: Normal (0) vs Pneumonia (1).
Pneumonia detection is a high-stakes clinical problem where missed positives (false
negatives) carry significant patient risk — making Recall a critical metric alongside AUC.

**Secondary — BreastMNIST**
Derived from the Dataset of Breast Ultrasound Images (BUSI). Binary task: Malignant (1)
vs Benign/Normal (0). Included to test cross-dataset generalisation of the RL policy.

### Preprocessing Applied by MedMNIST

- Images resized to 28 x 28 pixels (bicubic interpolation)
- Grayscale conversion (single channel)
- Official train / validation / test splits preserved
- Labels encoded as integers (0, 1, ...)

````python
# Dataset loading 
def get_transform() -> transforms.Compose:
    """Standard normalisation: pixel values in [-1, 1] after mean=0.5, std=0.5."""
    return transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.5], std=[0.5]),
    ])


def load_dataset(name: str, root: str = "./data") -> Tuple:
    """Download and load a MedMNIST sub-dataset."""
    info      = INFO[name]
    DataClass = getattr(medmnist, info["python_class"])
    tf        = get_transform()
    os.makedirs(root, exist_ok=True)

    t0 = time.time()
    train_ds = DataClass(split="train", transform=tf, download=True, root=root)
    val_ds   = DataClass(split="val",   transform=tf, download=True, root=root)
    test_ds  = DataClass(split="test",  transform=tf, download=True, root=root)
    elapsed  = time.time() - t0

    n_classes = len(info["label"])
    print(f"Dataset     : {name.upper()}")
    print(f"Task        : {info['task']}")
    print(f"Classes     : {n_classes}  ->  {info['label']}")
    print(f"Train / Val / Test : {len(train_ds)} / {len(val_ds)} / {len(test_ds)}")
    print(f"Load time   : {elapsed:.1f}s")
    return train_ds, val_ds, test_ds, n_classes, info


print("=" * 50)
print("PRIMARY DATASET")
print("=" * 50)
PRIMARY = "pneumoniamnist"
train_ds, val_ds, test_ds, N_CLASSES, DS_INFO = load_dataset(PRIMARY)

print()
print("=" * 50)
print("SECONDARY DATASET")
print("=" * 50)
SECONDARY = "breastmnist"
train_ds2, val_ds2, test_ds2, N_CLASSES2, DS_INFO2 = load_dataset(SECONDARY)
````

````python
# Extract raw numpy arrays 
def extract_arrays(dataset) -> Tuple[np.ndarray, np.ndarray]:
    """Convert a MedMNIST dataset to numpy arrays (images, labels)."""
    loader = DataLoader(dataset, batch_size=512, shuffle=False)
    imgs, lbls = [], []
    for images, labels in loader:
        imgs.append(images.numpy())
        lbls.append(labels.numpy().squeeze())
    return np.concatenate(imgs), np.concatenate(lbls)


# Primary
train_images, train_labels = extract_arrays(train_ds)
val_images,   val_labels   = extract_arrays(val_ds)
test_images,  test_labels  = extract_arrays(test_ds)

# Secondary
train_images2, train_labels2 = extract_arrays(train_ds2)
val_images2,   val_labels2   = extract_arrays(val_ds2)
test_images2,  test_labels2  = extract_arrays(test_ds2)

LABEL_NAMES  = DS_INFO["label"]
LABEL_NAMES2 = DS_INFO2["label"]

print(f"Primary ({PRIMARY.upper()}):")
print(f"  Train : {train_images.shape}  labels: {train_labels.shape}")
print(f"  Val   : {val_images.shape}    labels: {val_labels.shape}")
print(f"  Test  : {test_images.shape}   labels: {test_labels.shape}")
print()
print(f"Secondary ({SECONDARY.upper()}):")
print(f"  Train : {train_images2.shape}  labels: {train_labels2.shape}")
print(f"  Val   : {val_images2.shape}    labels: {val_labels2.shape}")
print(f"  Test  : {test_images2.shape}   labels: {test_labels2.shape}")
````

### 3.1 Dataset Summary Statistics

````python
# Descriptive statistics — both datasets
def compute_split_stats(images: np.ndarray, labels: np.ndarray,
                        label_names: Dict, split: str, dataset: str) -> Dict:
    """Compute and print per-split descriptive statistics. Returns a stats dict."""
    unique, counts = np.unique(labels, return_counts=True)
    imb_ratio = counts.max() / counts.min()
    flat = images.ravel()
    stats = {
        "dataset": dataset, "split": split,
        "n_samples": len(images),
        "n_classes": len(unique),
        "imbalance_ratio": imb_ratio,
        "pixel_mean": float(flat.mean()),
        "pixel_std":  float(flat.std()),
        "pixel_min":  float(flat.min()),
        "pixel_max":  float(flat.max()),
        "pixel_skew": float(skew(flat)),
        "pixel_kurtosis": float(kurtosis(flat)),
    }
    for cls, cnt in zip(unique, counts):
        cname = label_names.get(str(int(cls)), f"C{cls}")
        stats[f"class_{cls}_name"]  = cname
        stats[f"class_{cls}_count"] = int(cnt)
        stats[f"class_{cls}_pct"]   = round(100 * cnt / len(labels), 2)
    return stats


all_stats = []
print("=" * 72)
for ds_name, imgs_dict, lbl_names in [
    (PRIMARY,   {"Train": (train_images, train_labels),
                 "Val":   (val_images,   val_labels),
                 "Test":  (test_images,  test_labels)},   LABEL_NAMES),
    (SECONDARY, {"Train": (train_images2, train_labels2),
                 "Val":   (val_images2,   val_labels2),
                 "Test":  (test_images2,  test_labels2)}, LABEL_NAMES2),
]:
    print(f"\nDATASET: {ds_name.upper()}")
    print("-" * 72)
    print(f"  {'Split':>6} {'N':>6} {'Imbalance':>11} "
          f"{'Mean':>8} {'Std':>8} {'Skew':>8} {'Kurt':>8}")
    print("  " + "-" * 60)
    for split, (imgs, lbls) in imgs_dict.items():
        s = compute_split_stats(imgs, lbls, lbl_names, split, ds_name)
        all_stats.append(s)
        print(f"  {split:>6} {s['n_samples']:>6} {s['imbalance_ratio']:>11.2f} "
              f"{s['pixel_mean']:>8.4f} {s['pixel_std']:>8.4f} "
              f"{s['pixel_skew']:>8.4f} {s['pixel_kurtosis']:>8.4f}")
    print()
    # Per-class breakdown for train split
    s = all_stats[-3]  # train stats
    unique = np.unique(
        train_labels if ds_name == PRIMARY else train_labels2)
    imgs_arr = train_images if ds_name == PRIMARY else train_images2
    lbls_arr = train_labels if ds_name == PRIMARY else train_labels2
    for cls in unique:
        cname = lbl_names.get(str(int(cls)), f"Class {cls}")
        mask  = lbls_arr == cls
        pix   = imgs_arr[mask].ravel()
        print(f"    Class {cls} ({cname:20s}): n={mask.sum():5d} | "
              f"mean={pix.mean():.4f} std={pix.std():.4f} "
              f"skew={skew(pix):.3f}")

print("=" * 72)

# Save summary to CSV
df_stats = pd.DataFrame(all_stats)
csv_path = f"{EXPERIMENTS_DIR}/eda_dataset_stats.csv"
df_stats.to_csv(csv_path, index=False)
print(f"\nStats saved: {csv_path}")
````

### 3.3 Sample Image Grids

````python
# ── Sample images — one grid per class per dataset ─────────────────────────────

def plot_sample_grid(images, labels, label_names, dataset_name,
                     n_per_class=8, save_path=None):
    """
    Display a grid of n_per_class randomly selected images for each class.
    Images are shown in their normalised [-1, 1] range using a gray colormap.
    """
    unique_cls = np.unique(labels)
    n_cls      = len(unique_cls)
    fig, axes  = plt.subplots(n_cls, n_per_class,
                               figsize=(n_per_class * 1.5, n_cls * 2.0))
    if n_cls == 1:
        axes = axes[np.newaxis, :]
    fig.suptitle(f"Sample Images — {dataset_name.upper()} (Train Split)",
                 fontsize=13, fontweight="bold")

    rng = np.random.default_rng(SEED)
    for row, cls in enumerate(unique_cls):
        mask    = np.where(labels == cls)[0]
        chosen  = rng.choice(mask, n_per_class, replace=len(mask) < n_per_class)
        cname   = label_names.get(str(int(cls)), f"Class {cls}")
        for col, idx in enumerate(chosen):
            ax = axes[row, col]
            ax.imshow(images[idx].squeeze(), cmap="gray", vmin=-1, vmax=1)
            ax.axis("off")
            if col == 0:
                ax.set_ylabel(f"C{cls}\n{cname}", fontsize=9, fontweight="bold",
                              rotation=0, labelpad=50, va="center")
    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches="tight")
        print(f"Saved: {save_path}")
    plt.show()


plot_sample_grid(train_images, train_labels, LABEL_NAMES, PRIMARY,
                 save_path=f"{FIGURES_DIR}/eda_02_samples_{PRIMARY}.png")

plot_sample_grid(train_images2, train_labels2, LABEL_NAMES2, SECONDARY,
                 save_path=f"{FIGURES_DIR}/eda_02_samples_{SECONDARY}.png")
````

### 3.5 Per-Class Pixel Distributions

````python
# ── Per-class pixel distributions — train split ────────────────────────────────

def plot_per_class_pixel_dist(images, labels, label_names,
                               dataset_name, save_path=None):
    unique_cls = np.unique(labels)
    fig, axes  = plt.subplots(1, len(unique_cls),
                               figsize=(6 * len(unique_cls), 4))
    if len(unique_cls) == 1:
        axes = [axes]
    fig.suptitle(f"Per-Class Pixel Distribution — {dataset_name.upper()} (Train)",
                 fontsize=13, fontweight="bold")
    cls_stats_rows = []
    for ax, cls in zip(axes, unique_cls):
        mask   = labels == cls
        pixels = images[mask].ravel()
        cname  = label_names.get(str(int(cls)), f"Class {cls}")
        color  = COLORS[int(cls) % len(COLORS)]
        ax.hist(pixels, bins=60, color=color, alpha=0.80, density=True, edgecolor="none")
        ax.axvline(pixels.mean(), color="black", linestyle="--",
                   linewidth=1.8, label=f"mean={pixels.mean():.3f}")
        ax.axvline(np.median(pixels), color="gray", linestyle=":",
                   linewidth=1.5, label=f"median={np.median(pixels):.3f}")
        ax.set_title(f"Class {cls} — {cname}\n"
                     f"n={mask.sum()}, std={pixels.std():.3f}, "
                     f"skew={skew(pixels):.2f}",
                     fontsize=10, fontweight="bold")
        ax.set_xlabel("Pixel value"); ax.set_ylabel("Density")
        ax.legend(fontsize=9); ax.grid(alpha=0.35)
        cls_stats_rows.append({
            "dataset": dataset_name, "class": int(cls), "class_name": cname,
            "n": int(mask.sum()),
            "pixel_mean": float(pixels.mean()), "pixel_std": float(pixels.std()),
            "pixel_median": float(np.median(pixels)),
            "pixel_skew": float(skew(pixels)), "pixel_kurtosis": float(kurtosis(pixels)),
        })
    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches="tight")
        print(f"Saved: {save_path}")
    plt.show()
    return cls_stats_rows


rows1 = plot_per_class_pixel_dist(train_images,  train_labels,  LABEL_NAMES,
                                   PRIMARY,
                                   f"{FIGURES_DIR}/eda_04_perclass_pixels_{PRIMARY}.png")
rows2 = plot_per_class_pixel_dist(train_images2, train_labels2, LABEL_NAMES2,
                                   SECONDARY,
                                   f"{FIGURES_DIR}/eda_04_perclass_pixels_{SECONDARY}.png")

df_cls = pd.DataFrame(rows1 + rows2)
csv2   = f"{EXPERIMENTS_DIR}/eda_per_class_pixel_stats.csv"
df_cls.to_csv(csv2, index=False)
print(f"Per-class stats saved: {csv2}")
````

### 3.6 Mean and Standard-Deviation Images Per Class

````python
# ── Mean and std images — reveals class-discriminative spatial structure ────────

def plot_mean_std_images(images, labels, label_names,
                          dataset_name, save_path=None):
    """
    Mean image shows average spatial intensity per class.
    Std image shows intra-class variance — high std = model needs to focus there.
    """
    unique_cls = np.unique(labels)
    fig, axes  = plt.subplots(2, len(unique_cls),
                               figsize=(5 * len(unique_cls), 8))
    if len(unique_cls) == 1:
        axes = axes.reshape(2, 1)
    fig.suptitle(f"Mean and Std Images per Class — {dataset_name.upper()}",
                 fontsize=13, fontweight="bold")

    for col, cls in enumerate(unique_cls):
        mask     = labels == cls
        imgs_cls = images[mask]
        cname    = label_names.get(str(int(cls)), f"Class {cls}")
        mean_img = imgs_cls.mean(axis=0).squeeze()
        std_img  = imgs_cls.std(axis=0).squeeze()

        im0 = axes[0, col].imshow(mean_img, cmap="gray", vmin=-1, vmax=1)
        axes[0, col].set_title(f"Class {cls} — {cname}\n"
                                f"Mean image (n={mask.sum()})",
                                fontsize=10, fontweight="bold")
        axes[0, col].axis("off")
        plt.colorbar(im0, ax=axes[0, col], fraction=0.046, pad=0.04,
                     label="Normalised pixel value")

        im1 = axes[1, col].imshow(std_img, cmap="hot")
        axes[1, col].set_title(f"Class {cls} — Pixel Std\n"
                                f"(bright = high intra-class variance)",
                                fontsize=10)
        axes[1, col].axis("off")
        plt.colorbar(im1, ax=axes[1, col], fraction=0.046, pad=0.04,
                     label="Std dev")

    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches="tight")
        print(f"Saved: {save_path}")
    plt.show()


plot_mean_std_images(train_images,  train_labels,  LABEL_NAMES,
                     PRIMARY,
                     f"{FIGURES_DIR}/eda_05_mean_std_{PRIMARY}.png")
plot_mean_std_images(train_images2, train_labels2, LABEL_NAMES2,
                     SECONDARY,
                     f"{FIGURES_DIR}/eda_05_mean_std_{SECONDARY}.png")

print("Interpretation:")
print("  High std in the lung region (pneumonia) vs uniform low std (normal)")
print("  confirms that the pathology signal is spatially concentrated.")
print("  This justifies using a CNN rather than a pixel-average classifier.")
````

### 3.7 Image Quality — Sharpness and Contrast

````python
# ── Sharpness (Laplacian variance) and contrast (pixel std) ───────────────────
# Sharpness proxy: variance of the Laplacian operator applied to the image.
# Low sharpness = blurry image. High sharpness = fine edges present.
# Contrast proxy: per-image pixel standard deviation.

def compute_quality_metrics(images: np.ndarray, labels: np.ndarray,
                             n_sample: int = 2000, seed: int = SEED):
    rng     = np.random.default_rng(seed)
    idx     = rng.choice(len(images), min(n_sample, len(images)), replace=False)
    sharp   = np.array([float(laplace(images[i].squeeze()).var()) for i in idx])
    contrast = np.array([float(images[i].std()) for i in idx])
    return sharp, contrast, labels[idx], idx


def plot_quality(images, labels, label_names, dataset_name, save_path=None):
    sharp, contrast, s_lbls, _ = compute_quality_metrics(images, labels)
    unique_cls = np.unique(s_lbls)

    fig, axes = plt.subplots(1, 3, figsize=(16, 4))
    fig.suptitle(f"Image Quality Analysis — {dataset_name.upper()} (sampled n=2000)",
                 fontsize=13, fontweight="bold")

    # Sharpness histogram
    axes[0].hist(sharp, bins=50, color="#9C27B0", alpha=0.80, edgecolor="none", density=True)
    axes[0].axvline(np.median(sharp), color="black", linestyle="--",
                    linewidth=1.5, label=f"median={np.median(sharp):.3f}")
    axes[0].set_title("Sharpness (Laplacian variance)", fontweight="bold")
    axes[0].set_xlabel("Variance"); axes[0].set_ylabel("Density")
    axes[0].legend(fontsize=9); axes[0].grid(alpha=0.4)

    # Contrast histogram
    axes[1].hist(contrast, bins=50, color="#FF9800", alpha=0.80, edgecolor="none", density=True)
    axes[1].axvline(np.median(contrast), color="black", linestyle="--",
                    linewidth=1.5, label=f"median={np.median(contrast):.3f}")
    axes[1].set_title("Contrast (pixel std per image)", fontweight="bold")
    axes[1].set_xlabel("Pixel std"); axes[1].set_ylabel("Density")
    axes[1].legend(fontsize=9); axes[1].grid(alpha=0.4)

    # Scatter: contrast vs sharpness, coloured by class
    for cls in unique_cls:
        mask  = s_lbls == cls
        cname = label_names.get(str(int(cls)), f"C{cls}")
        axes[2].scatter(contrast[mask], sharp[mask],
                        c=COLORS[int(cls) % len(COLORS)],
                        alpha=0.25, s=8, label=f"C{cls}: {cname}")
    axes[2].set_title("Contrast vs Sharpness by Class", fontweight="bold")
    axes[2].set_xlabel("Contrast"); axes[2].set_ylabel("Sharpness")
    axes[2].legend(fontsize=9); axes[2].grid(alpha=0.3)

    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches="tight")
        print(f"Saved: {save_path}")
    plt.show()

    print(f"  Sharpness : mean={sharp.mean():.3f}  std={sharp.std():.3f}  "
          f"min={sharp.min():.3f}  max={sharp.max():.3f}")
    print(f"  Contrast  : mean={contrast.mean():.3f}  std={contrast.std():.3f}  "
          f"min={contrast.min():.3f}  max={contrast.max():.3f}")


print(f"PRIMARY — {PRIMARY.upper()}")
plot_quality(train_images, train_labels, LABEL_NAMES, PRIMARY,
             f"{FIGURES_DIR}/eda_06_image_quality_{PRIMARY}.png")
print()
print(f"SECONDARY — {SECONDARY.upper()}")
plot_quality(train_images2, train_labels2, LABEL_NAMES2, SECONDARY,
             f"{FIGURES_DIR}/eda_06_image_quality_{SECONDARY}.png")
````

### 3.8 Pixel Mean Separability Between Classes

````python
# ── Global pixel mean as a discriminability proxy ─────────────────────────────
# If classes are well-separated by mean pixel brightness alone, a linear model
# could suffice. Poor separation justifies the CNN feature extractor.

def plot_pixel_mean_separability(images, labels, label_names,
                                  dataset_name, save_path=None):
    pixel_means = images.mean(axis=(1, 2, 3))
    unique_cls  = np.unique(labels)

    fig, axes = plt.subplots(1, 2, figsize=(13, 4))
    fig.suptitle(f"Pixel Mean Separability — {dataset_name.upper()}",
                 fontsize=13, fontweight="bold")

    # Boxplot per class
    bp_data   = [pixel_means[labels == cls] for cls in unique_cls]
    bp_labels = [label_names.get(str(int(c)), f"C{c}") for c in unique_cls]
    bp = axes[0].boxplot(bp_data, labels=bp_labels, patch_artist=True, notch=True)
    for patch, color in zip(bp["boxes"], COLORS):
        patch.set_facecolor(color); patch.set_alpha(0.75)
    axes[0].set_title("Per-class pixel mean distribution", fontweight="bold")
    axes[0].set_ylabel("Mean pixel value")
    axes[0].grid(alpha=0.4, axis="y")

    # Overlapping histogram
    for cls in unique_cls:
        vals  = pixel_means[labels == cls]
        cname = label_names.get(str(int(cls)), f"C{cls}")
        axes[1].hist(vals, bins=50, color=COLORS[int(cls) % len(COLORS)],
                     alpha=0.55, density=True, label=f"C{cls}: {cname}", edgecolor="none")
    axes[1].set_title("Pixel mean distribution overlap", fontweight="bold")
    axes[1].set_xlabel("Mean pixel value"); axes[1].set_ylabel("Density")
    axes[1].legend(fontsize=9); axes[1].grid(alpha=0.4)

    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches="tight")
        print(f"Saved: {save_path}")
    plt.show()

    # Quantify overlap
    for cls in unique_cls:
        vals  = pixel_means[labels == cls]
        cname = label_names.get(str(int(cls)), f"C{cls}")
        print(f"    Class {cls} ({cname}): mean={vals.mean():.4f}  std={vals.std():.4f}")


print(f"PRIMARY — {PRIMARY.upper()}")
plot_pixel_mean_separability(train_images, train_labels, LABEL_NAMES, PRIMARY,
                              f"{FIGURES_DIR}/eda_07_pixel_mean_separability_{PRIMARY}.png")
print()
print(f"SECONDARY — {SECONDARY.upper()}")
plot_pixel_mean_separability(train_images2, train_labels2, LABEL_NAMES2, SECONDARY,
                              f"{FIGURES_DIR}/eda_07_pixel_mean_separability_{SECONDARY}.png")
print()
print("Interpretation: overlapping distributions confirm that pixel mean alone is")
print("insufficient — spatial features extracted by a CNN are necessary.")
````

### 3.9 Pixel Spatial Correlation

````python
# ── Pixel correlation matrix (subsampled 8x8 grid) ────────────────────────────
# High inter-pixel correlation confirms that spatial structure (not independent
# pixels) carries the signal, motivating convolutional rather than fully-connected
# feature extractors.

def plot_pixel_correlation(images, dataset_name, n_cov=1000, save_path=None):
    rng  = np.random.default_rng(SEED)
    idx  = rng.choice(len(images), min(n_cov, len(images)), replace=False)
    flat = images[idx].reshape(len(idx), -1)   # (N, 784)

    # Subsample to 8x8 spatial grid for readability
    step   = 28 // 8
    s_idx  = np.arange(28 * 28).reshape(28, 28)[::step, ::step].ravel()
    sub    = flat[:, s_idx]                    # (N, 64)
    corr   = np.corrcoef(sub.T)               # (64, 64)

    fig, ax = plt.subplots(figsize=(7, 5))
    im = ax.imshow(corr, cmap="RdBu_r", vmin=-1, vmax=1)
    plt.colorbar(im, ax=ax, label="Pearson correlation")
    ax.set_title(f"Pixel Correlation Matrix (8x8 grid) — {dataset_name.upper()}\n"
                 f"(sampled n={len(idx)}, step={step}px between sampled pixels)",
                 fontweight="bold")
    ax.set_xlabel("Spatial position index")
    ax.set_ylabel("Spatial position index")
    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches="tight")
        print(f"Saved: {save_path}")
    plt.show()

    off_diag = corr[np.triu_indices_from(corr, k=1)]
    print(f"  Off-diagonal correlations — mean: {off_diag.mean():.3f}  "
          f"std: {off_diag.std():.3f}  max: {off_diag.max():.3f}")
    print("  High positive correlations between neighbouring pixels confirm spatial")
    print("  structure that CNNs are designed to exploit.")


plot_pixel_correlation(train_images, PRIMARY,
                       save_path=f"{FIGURES_DIR}/eda_08_pixel_correlation_{PRIMARY}.png")
plot_pixel_correlation(train_images2, SECONDARY,
                       save_path=f"{FIGURES_DIR}/eda_08_pixel_correlation_{SECONDARY}.png")
````

### 3.10 Active Learning Pool Composition

````python
# ── Initial labeled pool for active learning setup ────────────────────────────
# The RL agent starts with a small labeled seed (100 images) and queries the
# annotation oracle to expand it. The pool composition affects early model quality
# and therefore the initial reward landscape.

INITIAL_LABELED = 100

def initialise_pools(n_total, n_labeled, seed=SEED):
    rng = np.random.default_rng(seed)
    idx = list(range(n_total))
    rng.shuffle(idx)
    return idx[:n_labeled], idx[n_labeled:]

labeled_pool, unlabeled_pool = initialise_pools(len(train_images), INITIAL_LABELED)
pool_labels = train_labels[labeled_pool]

print(f"Labeled pool   : {len(labeled_pool)} images ({100*len(labeled_pool)/len(train_images):.1f}% of train)")
print(f"Unlabeled pool : {len(unlabeled_pool)} images")
print(f"Annotation budget (later): 200 queries")
print()

# Visualise pool composition
unique_full, counts_full = np.unique(train_labels, return_counts=True)
unique_pool, counts_pool = np.unique(pool_labels, return_counts=True)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
fig.suptitle("Active Learning Pool Composition — Primary Dataset",
             fontsize=13, fontweight="bold")

# Full train vs pool proportion comparison
x = np.arange(len(unique_full)); w = 0.38
axes[0].bar(x - w/2, counts_full / counts_full.sum(), w,
            label="Full train set", color="#607D8B", alpha=0.8, edgecolor="black", linewidth=0.7)
axes[0].bar(x + w/2, counts_pool / counts_pool.sum(), w,
            label=f"Labeled pool (n={INITIAL_LABELED})",
            color="#2196F3", alpha=0.8, edgecolor="black", linewidth=0.7)
axes[0].set_xticks(x)
axes[0].set_xticklabels(
    [LABEL_NAMES.get(str(int(u)), f"C{u}") for u in unique_full])
axes[0].set_ylabel("Class proportion")
axes[0].set_title("Full train vs initial labeled pool", fontweight="bold")
axes[0].legend(); axes[0].grid(alpha=0.4, axis="y")

# Pie chart — labeled pool
axes[1].pie(counts_pool,
            labels=[LABEL_NAMES.get(str(int(u)), f"C{u}") for u in unique_pool],
            colors=COLORS[:len(unique_pool)],
            autopct="%1.1f%%", startangle=90, textprops={"fontsize": 11})
axes[1].set_title(f"Labeled pool class split\n(n={INITIAL_LABELED})",
                  fontweight="bold")

# Pie chart — unlabeled pool
u_unlab, c_unlab = np.unique(train_labels[unlabeled_pool], return_counts=True)
axes[2].pie(c_unlab,
            labels=[LABEL_NAMES.get(str(int(u)), f"C{u}") for u in u_unlab],
            colors=COLORS[:len(u_unlab)],
            autopct="%1.1f%%", startangle=90, textprops={"fontsize": 11})
axes[2].set_title(f"Unlabeled pool class split\n(n={len(unlabeled_pool)})",
                  fontweight="bold")

plt.tight_layout()
out = f"{FIGURES_DIR}/eda_09_pool_composition.png"
plt.savefig(out, dpi=150, bbox_inches="tight")
plt.show()
print(f"Saved: {out}")

# Imbalance in labeled pool
pool_imb = counts_pool.max() / counts_pool.min()
full_imb  = counts_full.max() / counts_full.min()
print(f"\nImbalance ratio — full train: {full_imb:.2f}x  |  labeled pool: {pool_imb:.2f}x")
print("Note: the small labeled pool may amplify imbalance, which biases early model")
print("predictions and therefore the entropy distribution seen by the RL agent.")
````

## 4. Baseline Models

This section implements and evaluates three baseline classifiers on the **full labeled training set**.
These baselines represent the performance ceiling achievable without any annotation budget constraints,
providing a reference point for the active learning experiment in later weeks.

| Baseline | Description | Purpose |
|----------|-------------|---------|
| Logistic Regression | Linear model on flat pixels | Sanity check — no spatial features |
| Custom CNN | 3-block ConvNet trained from scratch | Primary AL backbone — same architecture used by the RL agent |
| Frozen ResNet-18 | ImageNet pretrained, linear head only | Transfer learning upper bound |

**Why these metrics?**
- **AUC-ROC**: threshold-free discrimination; primary metric for class-imbalanced medical tasks
- **PR-AUC**: area under precision-recall curve; more informative than ROC when positives are rare  
- **Accuracy**: overall correctness
- **F1**: harmonic mean of precision and recall; balances both error types  
- **Recall (Sensitivity)**: fraction of true positives found; clinically critical (missed diagnosis = false negative)  
- **Precision**: fraction of predicted positives that are correct


````python
# ── DataLoader helpers ─────────────────────────────────────────────────────────

def make_loader(images, labels, indices=None, batch_size=64, shuffle=True) -> DataLoader:
    if indices is not None:
        imgs = torch.tensor(images[indices], dtype=torch.float32)
        lbs  = torch.tensor(labels[indices], dtype=torch.long)
    else:
        imgs = torch.tensor(images, dtype=torch.float32)
        lbs  = torch.tensor(labels, dtype=torch.long)
    return DataLoader(TensorDataset(imgs, lbs), batch_size=batch_size,
                      shuffle=shuffle, num_workers=0,
                      pin_memory=DEVICE.type == "cuda")

train_loader = make_loader(train_images, train_labels, shuffle=True)
val_loader   = make_loader(val_images,   val_labels,   shuffle=False)
test_loader  = make_loader(test_images,  test_labels,  shuffle=False)

print(f"Train loader : {len(train_loader)} batches x {train_loader.batch_size}")
print(f"Val loader   : {len(val_loader)} batches")
print(f"Test loader  : {len(test_loader)} batches")
````

````python
# ── Lightweight CNN (used in RL environment later) ─────────────────────────────

class MedicalCNN(nn.Module):
    """
    Lightweight 3-block CNN with BatchNorm and Dropout.
    Designed for 28x28 grayscale images and small labeled pools.
    Used as the classification backbone inside the RL active learning environment.
    """
    def __init__(self, n_classes: int, dropout_rate: float = 0.3, in_channels: int = 1):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(in_channels, 16, 3, padding=1), nn.BatchNorm2d(16),
            nn.ReLU(True), nn.MaxPool2d(2),          # 28 -> 14
            nn.Conv2d(16, 32, 3, padding=1), nn.BatchNorm2d(32),
            nn.ReLU(True), nn.MaxPool2d(2),          # 14 -> 7
            nn.Conv2d(32, 64, 3, padding=1), nn.BatchNorm2d(64),
            nn.ReLU(True), nn.AdaptiveAvgPool2d(4),  # -> 4x4
        )
        self.classifier = nn.Sequential(
            nn.Dropout(dropout_rate),
            nn.Linear(64 * 4 * 4, 128), nn.ReLU(True),
            nn.Dropout(dropout_rate / 2),
            nn.Linear(128, n_classes),
        )

    def forward(self, x):
        return self.classifier(self.features(x).flatten(1))


_dummy  = MedicalCNN(N_CLASSES).to(DEVICE)
_params = sum(p.numel() for p in _dummy.parameters() if p.requires_grad)
print(f"MedicalCNN   — trainable params: {_params:,}")
````

````python
# ── Frozen ResNet-18 feature extractor ────────────────────────────────────────

class FrozenResNet18(nn.Module):
    """
    ResNet-18 pre-trained on ImageNet with all convolutional layers frozen.
    Only the final linear classifier is trained.

    Adaptation for MedMNIST:
    - Input: 1-channel (grayscale) 28x28 images, replicated to 3 channels
    - The first conv layer accepts 3 channels (replicated from 1)
    - Feature dimension: 512 (ResNet-18 avgpool output)
    """
    def __init__(self, n_classes: int, freeze_backbone: bool = True):
        super().__init__()
        backbone = tv_models.resnet18(weights=tv_models.ResNet18_Weights.DEFAULT)

        # Replace first conv to handle 28x28 (stride=1, no aggressive downsampling)
        backbone.conv1 = nn.Conv2d(3, 64, kernel_size=3, stride=1, padding=1, bias=False)
        backbone.maxpool = nn.Identity()  # skip maxpool — images are small

        self.backbone = nn.Sequential(*list(backbone.children())[:-1])  # drop fc
        self.head     = nn.Linear(512, n_classes)

        if freeze_backbone:
            for param in self.backbone.parameters():
                param.requires_grad = False

        frozen = sum(p.numel() for p in self.backbone.parameters() if not p.requires_grad)
        trainable = sum(p.numel() for p in self.parameters() if p.requires_grad)
        print(f"FrozenResNet18 — frozen: {frozen:,}  trainable: {trainable:,}")

    def forward(self, x):
        # Replicate grayscale to 3 channels
        if x.shape[1] == 1:
            x = x.repeat(1, 3, 1, 1)
        feat = self.backbone(x).flatten(1)
        return self.head(feat)


resnet_model = FrozenResNet18(N_CLASSES, freeze_backbone=True).to(DEVICE)
````

````python
# ── Training and evaluation utilities ─────────────────────────────────────────

def compute_class_weights(labels: np.ndarray) -> torch.Tensor:
    """Inverse-frequency class weights to handle imbalance."""
    unique, counts = np.unique(labels, return_counts=True)
    weights = 1.0 / counts.astype(float)
    weights = weights / weights.sum() * len(unique)
    return torch.tensor(weights, dtype=torch.float32).to(DEVICE)


def train_one_epoch(model, loader, optimizer, criterion) -> float:
    model.train()
    total_loss = 0.0
    for images, labels in loader:
        images, labels = images.to(DEVICE), labels.to(DEVICE)
        optimizer.zero_grad()
        loss = criterion(model(images), labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    return total_loss / len(loader)


def evaluate_model(model, loader, n_classes) -> Dict:
    """Full evaluation: AUC, accuracy, F1, recall, precision, PR-AUC."""
    model.eval()
    all_probs, all_labels = [], []
    with torch.no_grad():
        for images, labels in loader:
            p = F.softmax(model(images.to(DEVICE)), dim=-1).cpu().numpy()
            all_probs.append(p)
            all_labels.append(labels.numpy())
    probs  = np.concatenate(all_probs)
    labels = np.concatenate(all_labels)
    preds  = probs.argmax(axis=1)
    avg    = "binary" if n_classes == 2 else "macro"
    try:
        auc = (roc_auc_score(labels, probs[:, 1])
               if n_classes == 2
               else roc_auc_score(labels, probs, multi_class="ovr", average="macro"))
    except ValueError:
        auc = 0.5
    try:
        pr_auc = average_precision_score(labels, probs[:, 1]) if n_classes == 2 else 0.0
    except Exception:
        pr_auc = 0.0
    return {
        "auc": auc, "pr_auc": pr_auc,
        "acc": accuracy_score(labels, preds),
        "f1":  f1_score(labels, preds, average=avg, zero_division=0),
        "recall":    recall_score(labels, preds, average=avg, zero_division=0),
        "precision": precision_score(labels, preds, average=avg, zero_division=0),
        "probs": probs, "labels": labels, "preds": preds,
    }


def train_model(model, train_loader, val_loader, n_classes,
                epochs=20, lr=1e-3, wd=1e-4,
                use_class_weights=True, verbose=True) -> Dict:
    """
    Training loop with cosine LR schedule, early stopping (patience=5),
    and best-model checkpointing by validation AUC.
    """
    all_labels = np.concatenate([
        lbls.numpy() for _, lbls in train_loader
    ])
    criterion = (nn.CrossEntropyLoss(weight=compute_class_weights(all_labels))
                 if use_class_weights else nn.CrossEntropyLoss())
    optimizer = optim.Adam(
        filter(lambda p: p.requires_grad, model.parameters()),
        lr=lr, weight_decay=wd)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs)

    best_auc, best_state, patience, history = 0.0, None, 0, []
    for epoch in range(1, epochs + 1):
        loss = train_one_epoch(model, train_loader, optimizer, criterion)
        scheduler.step()
        m = evaluate_model(model, val_loader, n_classes)
        history.append({"epoch": epoch, "loss": loss,
                        "val_auc": m["auc"], "val_acc": m["acc"],
                        "val_f1": m["f1"], "val_recall": m["recall"]})
        if m["auc"] > best_auc:
            best_auc  = m["auc"]
            best_state = {k: v.clone() for k, v in model.state_dict().items()}
            patience  = 0
        else:
            patience += 1
        if verbose:
            print(f"  Epoch {epoch:3d}/{epochs} | Loss: {loss:.4f} | "
                  f"Val AUC: {m['auc']:.4f} | Acc: {m['acc']:.4f} | "
                  f"F1: {m['f1']:.4f} | Recall: {m['recall']:.4f}")
        if patience >= 5:
            if verbose:
                print(f"  Early stopping at epoch {epoch} (best Val AUC={best_auc:.4f})")
            break
    if best_state:
        model.load_state_dict(best_state)
    return {"history": history, "best_val_auc": best_auc}


print("Training utilities defined.")
````

````python
# ── Baseline 1: Logistic Regression on flat pixels ────────────────────────────
print("=" * 60)
print("BASELINE 1 — Logistic Regression (flat pixels)")
print("=" * 60)
set_seed(SEED)

X_train = train_images.reshape(len(train_images), -1)
X_val   = val_images.reshape(len(val_images),   -1)
X_test  = test_images.reshape(len(test_images),  -1)

scaler  = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val   = scaler.transform(X_val)
X_test  = scaler.transform(X_test)

lr_model = LogisticRegression(max_iter=1000, C=1.0, class_weight="balanced",
                               random_state=SEED, solver="lbfgs")
t0 = time.time()
lr_model.fit(X_train, train_labels)
print(f"  Training time: {time.time()-t0:.1f}s")

def evaluate_sklearn(model, X, y, n_classes=2) -> Dict:
    preds = model.predict(X)
    probs = model.predict_proba(X)
    avg   = "binary" if n_classes == 2 else "macro"
    auc   = roc_auc_score(y, probs[:, 1]) if n_classes == 2 else 0.0
    pr_auc = average_precision_score(y, probs[:, 1]) if n_classes == 2 else 0.0
    return {
        "auc": auc, "pr_auc": pr_auc,
        "acc": accuracy_score(y, preds),
        "f1":  f1_score(y, preds, average=avg, zero_division=0),
        "recall":    recall_score(y, preds, average=avg, zero_division=0),
        "precision": precision_score(y, preds, average=avg, zero_division=0),
        "probs": probs, "labels": y, "preds": preds,
    }

lr_val  = evaluate_sklearn(lr_model, X_val,  val_labels)
lr_test = evaluate_sklearn(lr_model, X_test, test_labels)
print(f"  Val  — AUC:{lr_val['auc']:.4f}  Acc:{lr_val['acc']:.4f}  "
      f"F1:{lr_val['f1']:.4f}  Recall:{lr_val['recall']:.4f}")
print(f"  Test — AUC:{lr_test['auc']:.4f}  Acc:{lr_test['acc']:.4f}  "
      f"F1:{lr_test['f1']:.4f}  Recall:{lr_test['recall']:.4f}")
````

````python
# ── Baseline 2: Custom CNN trained from scratch (full train set) ───────────────
print("=" * 60)
print("BASELINE 2 — Custom CNN (trained from scratch)")
print("=" * 60)
set_seed(SEED)

cnn_model = MedicalCNN(N_CLASSES, dropout_rate=0.3).to(DEVICE)
t0 = time.time()
cnn_hist  = train_model(cnn_model, train_loader, val_loader, N_CLASSES,
                         epochs=25, lr=1e-3, verbose=True)
print(f"\n  Training time: {time.time()-t0:.1f}s")

cnn_val  = evaluate_model(cnn_model, val_loader,  N_CLASSES)
cnn_test = evaluate_model(cnn_model, test_loader, N_CLASSES)
print(f"\n  Val  — AUC:{cnn_val['auc']:.4f}  Acc:{cnn_val['acc']:.4f}  "
      f"F1:{cnn_val['f1']:.4f}  Recall:{cnn_val['recall']:.4f}")
print(f"  Test — AUC:{cnn_test['auc']:.4f}  Acc:{cnn_test['acc']:.4f}  "
      f"F1:{cnn_test['f1']:.4f}  Recall:{cnn_test['recall']:.4f}")
````

````python
# ── Baseline 3: Frozen ResNet-18 (imaging SOTA baseline) ──────────────────────
print("=" * 60)
print("BASELINE 3 — Frozen ResNet-18 + Linear Head  [primary baseline]")
print("=" * 60)
set_seed(SEED)

t0 = time.time()
resnet_hist = train_model(resnet_model, train_loader, val_loader, N_CLASSES,
                           epochs=25, lr=1e-3, verbose=True)
print(f"\n  Training time: {time.time()-t0:.1f}s")

resnet_val  = evaluate_model(resnet_model, val_loader,  N_CLASSES)
resnet_test = evaluate_model(resnet_model, test_loader, N_CLASSES)
print(f"\n  Val  — AUC:{resnet_val['auc']:.4f}  Acc:{resnet_val['acc']:.4f}  "
      f"F1:{resnet_val['f1']:.4f}  Recall:{resnet_val['recall']:.4f}")
print(f"  Test — AUC:{resnet_test['auc']:.4f}  Acc:{resnet_test['acc']:.4f}  "
      f"F1:{resnet_test['f1']:.4f}  Recall:{resnet_test['recall']:.4f}")
````

### 4.9  Active Learning Baseline Metrics

The cells below run the **Active Learning strategies** (Random, Margin, Entropy, BALD) and report:

| Metric | Definition | Relevance |
|--------|-----------|-----------|
| **Final Val AUC** | AUC after all budget queries | End-point quality |
| **Final Test AUC** | AUC on held-out test set | Generalisation |
| **F1 / Recall** | Per-strategy clinical performance | Medical relevance |
| **ALC** | Area under the learning curve = mean Val AUC over all steps | Measures *how fast* a strategy learns — not just its final AUC |
| **Annotation Efficiency** | (Final AUC − Initial AUC) / Queries | AUC gain per label spent |

These are the metrics the RL agent will be benchmarked against in Week 3.


````python
# ── Run AL baselines (Random, Margin, Entropy, BALD) ──────────────────────────
# These strategies are run here as W2 baselines.
# The RL agent (DQN/Double DQN/Dueling DQN) will be trained and compared in W3.
import copy
from tqdm import tqdm

def compute_entropy_al(probs, eps=1e-10):
    return -np.sum(probs * np.log(probs + eps), axis=1)

def get_predictions_al(model, images, indices, batch_size=256):
    model.eval()
    loader = DataLoader(
        TensorDataset(torch.tensor(images[indices], dtype=torch.float32)),
        batch_size=batch_size, shuffle=False)
    all_probs = []
    with torch.no_grad():
        for (batch,) in loader:
            all_probs.append(F.softmax(model(batch.to(DEVICE)), dim=-1).cpu().numpy())
    probs_arr = np.concatenate(all_probs)
    return probs_arr, compute_entropy_al(probs_arr)

def random_query_al(unlabeled, n):
    return random.sample(unlabeled, min(n, len(unlabeled)))

def entropy_query_al(model, images, unlabeled, n):
    _, entropies = get_predictions_al(model, images, unlabeled)
    ranked = np.argsort(-entropies)[:n]
    return [unlabeled[i] for i in ranked]

def margin_query_al(model, images, unlabeled, n):
    probs, _ = get_predictions_al(model, images, unlabeled)
    sorted_p = np.sort(probs, axis=1)[:, ::-1]
    margins  = sorted_p[:, 0] - sorted_p[:, 1]
    ranked   = np.argsort(margins)[:n]
    return [unlabeled[i] for i in ranked]

def enable_dropout_al(model):
    model.eval()
    for m in model.modules():
        if isinstance(m, nn.Dropout): m.train()

def bald_query_al(model, images, unlabeled, n, T=5, batch_size=128):
    enable_dropout_al(model)
    loader = DataLoader(
        TensorDataset(torch.tensor(images[unlabeled], dtype=torch.float32)),
        batch_size=batch_size, shuffle=False)
    passes = []
    for _ in range(T):
        pp = []
        for (b,) in loader:
            with torch.no_grad():
                pp.append(F.softmax(model(b.to(DEVICE)), dim=-1).cpu().numpy())
        passes.append(np.concatenate(pp))
    passes = np.stack(passes)
    mean_p = passes.mean(0)
    scores = compute_entropy_al(mean_p) - np.mean([compute_entropy_al(p) for p in passes], 0)
    model.eval()
    ranked = np.argsort(-scores)[:n]
    return [unlabeled[i] for i in ranked]

def train_model_al(model, train_loader, val_loader, n_classes, epochs=15, lr=1e-3, verbose=False):
    optimizer = optim.Adam(model.parameters(), lr=lr, weight_decay=1e-4)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs)
    criterion = nn.CrossEntropyLoss()
    best_auc, best_state, patience = 0.0, None, 0
    for epoch in range(epochs):
        model.train()
        for imgs, lbls in train_loader:
            imgs, lbls = imgs.to(DEVICE), lbls.to(DEVICE)
            optimizer.zero_grad()
            criterion(model(imgs), lbls).backward()
            optimizer.step()
        model.eval()
        all_p, all_l = [], []
        with torch.no_grad():
            for imgs, lbls in val_loader:
                all_p.append(F.softmax(model(imgs.to(DEVICE)), dim=-1).cpu().numpy())
                all_l.append(lbls.numpy())
        pp = np.concatenate(all_p); ll = np.concatenate(all_l)
        auc = roc_auc_score(ll, pp[:, 1]) if n_classes == 2 else 0.5
        scheduler.step()
        if auc > best_auc:
            best_auc = auc; best_state = copy.deepcopy(model.state_dict()); patience = 0
        else:
            patience += 1
        if patience >= 5: break
    if best_state: model.load_state_dict(best_state)
    return best_auc

def evaluate_al(model, loader, n_classes):
    model.eval()
    all_p, all_l = [], []
    with torch.no_grad():
        for imgs, lbls in loader:
            all_p.append(F.softmax(model(imgs.to(DEVICE)), dim=-1).cpu().numpy())
            all_l.append(lbls.numpy())
    pp = np.concatenate(all_p); ll = np.concatenate(all_l)
    preds = pp.argmax(1)
    avg = 'binary' if n_classes == 2 else 'macro'
    auc = roc_auc_score(ll, pp[:, 1]) if n_classes == 2 else 0.5
    return {
        'auc': auc,
        'acc': accuracy_score(ll, preds),
        'f1':  f1_score(ll, preds, average=avg, zero_division=0),
        'recall': recall_score(ll, preds, average=avg, zero_division=0),
        'precision': precision_score(ll, preds, average=avg, zero_division=0),
        'probs': pp, 'labels': ll, 'preds': preds,
    }

def run_al_strategy(strategy, train_images, train_labels,
                    val_loader, test_loader, n_classes,
                    n_steps, budget, init_labeled, init_unlabeled, q_size=10):
    set_seed(SEED)
    labeled   = list(init_labeled)
    unlabeled = list(init_unlabeled)
    remaining = budget
    queries_made, val_aucs, test_aucs, n_labels_hist = [0], [], [], []
    total_q = 0

    model = MedicalCNN(n_classes, 0.3).to(DEVICE)
    loader = make_loader(train_images, train_labels, labeled, 64)
    train_model_al(model, loader, val_loader, n_classes)
    val_aucs.append(evaluate_al(model, val_loader,  n_classes)['auc'])
    test_aucs.append(evaluate_al(model, test_loader, n_classes)['auc'])
    n_labels_hist.append(len(labeled))

    for _ in tqdm(range(n_steps), desc=f"{strategy:10s}", leave=False):
        if remaining <= 0 or not unlabeled: break
        q = min(q_size, remaining, len(unlabeled))
        if   strategy == 'random':  queried = random_query_al(unlabeled, q)
        elif strategy == 'entropy': queried = entropy_query_al(model, train_images, unlabeled, q)
        elif strategy == 'margin':  queried = margin_query_al(model, train_images, unlabeled, q)
        elif strategy == 'bald':    queried = bald_query_al(model, train_images, unlabeled, q)
        else:                       queried = random_query_al(unlabeled, q)

        for idx in queried: unlabeled.remove(idx)
        labeled.extend(queried); remaining -= q; total_q += q

        model  = MedicalCNN(n_classes, 0.3).to(DEVICE)
        loader = make_loader(train_images, train_labels, labeled, 64)
        train_model_al(model, loader, val_loader, n_classes)
        v = evaluate_al(model, val_loader,  n_classes)['auc']
        t = evaluate_al(model, test_loader, n_classes)['auc']
        queries_made.append(total_q); val_aucs.append(v); test_aucs.append(t)
        n_labels_hist.append(len(labeled))

    return {'strategy': strategy, 'queries': queries_made,
            'val_aucs': val_aucs, 'test_aucs': test_aucs,
            'n_labels': n_labels_hist, 'final_model': model}

# ── Run all 4 AL strategies ───────────────────────────────────────────────────
NUM_AL_STEPS   = 20
ANNOTATION_BUD = 200
QUERY_SIZE     = 10

print("Running AL baselines... (this takes several minutes)")
al_random  = run_al_strategy('random',  train_images, train_labels, val_loader, test_loader,
                              N_CLASSES, NUM_AL_STEPS, ANNOTATION_BUD,
                              labeled_pool, unlabeled_pool, QUERY_SIZE)
print(f"  Random  done — Final Val AUC: {al_random['val_aucs'][-1]:.4f}")

al_margin  = run_al_strategy('margin',  train_images, train_labels, val_loader, test_loader,
                              N_CLASSES, NUM_AL_STEPS, ANNOTATION_BUD,
                              labeled_pool, unlabeled_pool, QUERY_SIZE)
print(f"  Margin  done — Final Val AUC: {al_margin['val_aucs'][-1]:.4f}")

al_entropy = run_al_strategy('entropy', train_images, train_labels, val_loader, test_loader,
                              N_CLASSES, NUM_AL_STEPS, ANNOTATION_BUD,
                              labeled_pool, unlabeled_pool, QUERY_SIZE)
print(f"  Entropy done — Final Val AUC: {al_entropy['val_aucs'][-1]:.4f}")

al_bald    = run_al_strategy('bald',    train_images, train_labels, val_loader, test_loader,
                              N_CLASSES, NUM_AL_STEPS, ANNOTATION_BUD,
                              labeled_pool, unlabeled_pool, QUERY_SIZE)
print(f"  BALD    done — Final Val AUC: {al_bald['val_aucs'][-1]:.4f}")

al_results = [al_random, al_margin, al_entropy, al_bald]
print("\nAll AL strategies complete.")
````

## 5. Baseline CNN Classifier

````python
class MedicalCNN(nn.Module):
    """Lightweight CNN with 3 conv blocks + dropout for small labeled pools."""
    def __init__(self, n_classes: int, dropout_rate: float = 0.3, in_channels: int = 1):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(in_channels, 16, 3, padding=1), nn.BatchNorm2d(16), nn.ReLU(True), nn.MaxPool2d(2),
            nn.Conv2d(16, 32, 3, padding=1), nn.BatchNorm2d(32), nn.ReLU(True), nn.MaxPool2d(2),
            nn.Conv2d(32, 64, 3, padding=1), nn.BatchNorm2d(64), nn.ReLU(True), nn.AdaptiveAvgPool2d(4),
        )
        self.classifier = nn.Sequential(
            nn.Dropout(dropout_rate),
            nn.Linear(64 * 4 * 4, 128), nn.ReLU(True),
            nn.Dropout(dropout_rate / 2),
            nn.Linear(128, n_classes),
        )

    def forward(self, x):
        return self.classifier(self.features(x).flatten(1))

    def predict_proba(self, x):
        with torch.no_grad():
            return F.softmax(self.forward(x), dim=-1)


dummy = MedicalCNN(N_CLASSES).to(DEVICE)
out = dummy(torch.randn(4, 1, 28, 28).to(DEVICE))
print(f"CNN output: {out.shape}  |  Params: {sum(p.numel() for p in dummy.parameters()):,}")
````

````python
def train_epoch(model, loader, optimizer, criterion) -> float:
    model.train()
    total_loss = 0.0
    for images, labels in loader:
        images, labels = images.to(DEVICE), labels.to(DEVICE)
        optimizer.zero_grad()
        loss = criterion(model(images), labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    return total_loss / len(loader)


def evaluate(model, loader, n_classes) -> dict:
    model.eval()
    all_probs, all_labels = [], []
    with torch.no_grad():
        for images, labels in loader:
            probs = F.softmax(model(images.to(DEVICE)), dim=-1).cpu().numpy()
            all_probs.append(probs)
            all_labels.append(labels.numpy())
    probs_arr  = np.concatenate(all_probs)
    labels_arr = np.concatenate(all_labels)
    preds_arr  = probs_arr.argmax(axis=1)
    acc = accuracy_score(labels_arr, preds_arr)
    if n_classes == 2:
        auc = roc_auc_score(labels_arr, probs_arr[:, 1])
    else:
        try:
            auc = roc_auc_score(labels_arr, probs_arr, multi_class='ovr', average='macro')
        except ValueError:
            auc = 0.5
    avg = 'binary' if n_classes == 2 else 'macro'
    return {
        'acc': acc, 'auc': auc,
        'f1': f1_score(labels_arr, preds_arr, average=avg, zero_division=0),
        'recall': recall_score(labels_arr, preds_arr, average=avg, zero_division=0),
        'precision': precision_score(labels_arr, preds_arr, average=avg, zero_division=0),
    }


def train_model(model, train_loader, val_loader, n_classes, epochs=None, lr=None, verbose=False) -> float:
    epochs = epochs or CONFIG['cnn_epochs']
    lr     = lr     or CONFIG['cnn_lr']
    optimizer = optim.Adam(model.parameters(), lr=lr, weight_decay=CONFIG['cnn_weight_decay'])
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs)
    criterion = nn.CrossEntropyLoss()
    best_auc, best_state, patience_ctr = 0.0, None, 0
    for epoch in range(epochs):
        loss = train_epoch(model, train_loader, optimizer, criterion)
        auc = evaluate(model, val_loader, n_classes)['auc']
        scheduler.step()
        if auc > best_auc:
            best_auc = auc
            best_state = copy.deepcopy(model.state_dict())
            patience_ctr = 0
        else:
            patience_ctr += 1
        if verbose:
            print(f"  Epoch {epoch+1:3d}/{epochs} | Loss: {loss:.4f} | Val AUC: {auc:.4f}")
        if patience_ctr >= 5:
            if verbose: print(f" Early stopping at epoch {epoch+1}")
            break
    if best_state:
        model.load_state_dict(best_state)
    return best_auc


print("Training utilities defined.")
````

````python
print("Training initial CNN on labeled pool...")
set_seed(SEED)

init_model = MedicalCNN(N_CLASSES, CONFIG['dropout_rate']).to(DEVICE)
init_loader = make_loader(train_images, train_labels, labeled_pool, CONFIG['batch_size'])
init_auc = train_model(init_model, init_loader, val_loader, N_CLASSES, verbose=True)

m_test = evaluate(init_model, test_loader, N_CLASSES)
print(f"\nInitial Model (n={len(labeled_pool)} labels):")
print(f"   Val AUC: {init_auc:.4f}  |  Test AUC: {m_test['auc']:.4f}  |  Recall: {m_test['recall']:.4f}")
````

## 6. Uncertainty Measures & AL Baselines

We implement four query strategies:
- **Random**: uniform sampling (baseline)
- **Entropy** H(p) = −Σ pᵢ log pᵢ  (myopic uncertainty)
- **Margin** = p₁ − p₂  (smallest decision margin)
- **BALD** ≈ H[mean probs] − mean[H[per-pass probs]]  (epistemic uncertainty via MC-Dropout)


````python
def compute_entropy(probs: np.ndarray, eps: float = 1e-10) -> np.ndarray:
    """Shannon entropy H(p) = -sum(p_i * log(p_i))."""
    return -np.sum(probs * np.log(probs + eps), axis=1)


def get_predictions(model, images, indices, batch_size=256):
    """Batch inference → (probs [N,C], entropy [N])."""
    model.eval()
    loader = DataLoader(TensorDataset(torch.tensor(images[indices], dtype=torch.float32)),
                        batch_size=batch_size, shuffle=False)
    all_probs = []
    with torch.no_grad():
        for (batch,) in loader:
            all_probs.append(F.softmax(model(batch.to(DEVICE)), dim=-1).cpu().numpy())
    probs_arr = np.concatenate(all_probs)
    return probs_arr, compute_entropy(probs_arr)


def random_query(unlabeled_pool, n_query):
    return random.sample(unlabeled_pool, min(n_query, len(unlabeled_pool)))


def uncertainty_query(model, images, unlabeled_pool, n_query):
    _, entropies = get_predictions(model, images, unlabeled_pool)
    ranked = np.argsort(-entropies)[:n_query]
    return [unlabeled_pool[i] for i in ranked]


def margin_sampling_query(model, images, unlabeled_pool, n_query):
    probs, _ = get_predictions(model, images, unlabeled_pool)
    sorted_p = np.sort(probs, axis=1)[:, ::-1]
    margins  = sorted_p[:, 0] - sorted_p[:, 1]
    ranked   = np.argsort(margins)[:n_query]
    return [unlabeled_pool[i] for i in ranked]


def enable_dropout(model):
    model.eval()
    for m in model.modules():
        if isinstance(m, nn.Dropout):
            m.train()


def bald_query(model, images, unlabeled_pool, n_query, T=10, batch_size=128):
    """MC-Dropout BALD: I[y; w | x, D] ≈ H(mean) - mean(H per pass)."""
    enable_dropout(model)
    loader = DataLoader(TensorDataset(torch.tensor(images[unlabeled_pool], dtype=torch.float32)),
                        batch_size=batch_size, shuffle=False)
    all_passes = []
    for _ in range(T):
        pass_probs = []
        for (batch,) in loader:
            with torch.no_grad():
                pass_probs.append(F.softmax(model(batch.to(DEVICE)), dim=-1).cpu().numpy())
        all_passes.append(np.concatenate(pass_probs))
    all_passes = np.stack(all_passes)        # [T, N, C]
    mean_probs = all_passes.mean(axis=0)     # [N, C]
    bald_scores = compute_entropy(mean_probs) - np.mean(
        [compute_entropy(p) for p in all_passes], axis=0)
    model.eval()
    ranked = np.argsort(-bald_scores)[:n_query]
    return [unlabeled_pool[i] for i in ranked]


print(" Entropy, Margin, BALD query strategies defined.")
# Quick sanity check
u = np.array([[0.5, 0.5], [0.9, 0.1]])
print(f"   Uniform entropy: {compute_entropy(u)[0]:.4f}  Confident: {compute_entropy(u)[1]:.4f}")
````

````python
def run_active_learning(
    strategy: str,
    train_images, train_labels, val_loader, test_loader,
    n_classes, n_steps, budget, initial_labeled, initial_unlabeled,
    query_size: int = 10
) -> Dict:
    """Generic AL loop supporting random, entropy, margin, bald strategies."""
    set_seed(SEED)
    labeled   = list(initial_labeled)
    unlabeled = list(initial_unlabeled)
    remaining = budget
    queries_made, val_aucs, test_aucs, n_labels = [0], [], [], []
    total_q = 0

    model = MedicalCNN(n_classes, CONFIG['dropout_rate']).to(DEVICE)
    loader = make_loader(train_images, train_labels, labeled, CONFIG['batch_size'])
    train_model(model, loader, val_loader, n_classes)
    val_aucs.append(evaluate(model, val_loader,  n_classes)['auc'])
    test_aucs.append(evaluate(model, test_loader, n_classes)['auc'])
    n_labels.append(len(labeled))

    for _ in tqdm(range(n_steps), desc=f"{strategy:10s}", leave=False):
        if remaining <= 0 or not unlabeled: break
        actual_q = min(query_size, remaining, len(unlabeled))
        if strategy == 'random':
            queried = random_query(unlabeled, actual_q)
        elif strategy == 'entropy':
            queried = uncertainty_query(model, train_images, unlabeled, actual_q)
        elif strategy == 'margin':
            queried = margin_sampling_query(model, train_images, unlabeled, actual_q)
        elif strategy == 'bald':
            queried = bald_query(model, train_images, unlabeled, actual_q, T=5)
        else:
            queried = random_query(unlabeled, actual_q)
        for idx in queried: unlabeled.remove(idx)
        labeled.extend(queried)
        remaining -= actual_q
        total_q   += actual_q
        model = MedicalCNN(n_classes, CONFIG['dropout_rate']).to(DEVICE)
        loader = make_loader(train_images, train_labels, labeled, CONFIG['batch_size'])
        train_model(model, loader, val_loader, n_classes)
        v = evaluate(model, val_loader,  n_classes)['auc']
        t = evaluate(model, test_loader, n_classes)['auc']
        queries_made.append(total_q); val_aucs.append(v); test_aucs.append(t); n_labels.append(len(labeled))

    return {'strategy': strategy, 'queries': queries_made, 'val_aucs': val_aucs,
            'test_aucs': test_aucs, 'n_labels': n_labels, 'final_model': model}


print(" Running AL baselines...")
results_random  = run_active_learning('random',  train_images, train_labels, val_loader, test_loader,
                                      N_CLASSES, CONFIG['num_al_steps'], CONFIG['annotation_budget'],
                                      labeled_pool, unlabeled_pool, CONFIG['query_batch_size'])
print(f"1] Random  — Final Val AUC: {results_random['val_aucs'][-1]:.4f}")

results_margin  = run_active_learning('margin',  train_images, train_labels, val_loader, test_loader,
                                      N_CLASSES, CONFIG['num_al_steps'], CONFIG['annotation_budget'],
                                      labeled_pool, unlabeled_pool, CONFIG['query_batch_size'])
print(f"2] Margin  — Final Val AUC: {results_margin['val_aucs'][-1]:.4f}")

results_entropy = run_active_learning('entropy', train_images, train_labels, val_loader, test_loader,
                                      N_CLASSES, CONFIG['num_al_steps'], CONFIG['annotation_budget'],
                                      labeled_pool, unlabeled_pool, CONFIG['query_batch_size'])
print(f"3] Entropy — Final Val AUC: {results_entropy['val_aucs'][-1]:.4f}")

results_bald    = run_active_learning('bald',    train_images, train_labels, val_loader, test_loader,
                                      N_CLASSES, CONFIG['num_al_steps'], CONFIG['annotation_budget'],
                                      labeled_pool, unlabeled_pool, CONFIG['query_batch_size'])
print(f"4] BALD    — Final Val AUC: {results_bald['val_aucs'][-1]:.4f}")
````

## 7. RL Environment

### The 8-dim state
| Feature | Why    |
|---------|-------------|
| `confidence` | max p — how sure the model is |
| `entropy` | H(p) — uncertainty signal |
| `margin` | p₁−p₂ — how close the top-2 predictions are |
| `budget_ratio` | remaining budget fraction — teaches budget conservation |
| `auc_norm` | current model quality |
| `labeled_ratio` | pool maturity — diminishing returns detection |
| `step_progress` | 1 − budget_ratio = how far into the episode |
| `imbalance` | min_count/max_count in labeled pool — diversity signal |

### Shaped Reward Design
```
r = 0.5 × entropy + label_penalty        (proxy reward at annotation time)
r += 10 × ΔAUC                           (actual improvement bonus)
r = 0.01 × confidence                    (tiny reward for confident skips)
```
This gives the agent **dense, informative feedback** at every step, not just when retraining occurs.


````python
class AnnotationEnvironment:
    """
    SOTA RL environment for adaptive medical image annotation.

    """

    def __init__(self, train_images, train_labels, val_loader,
                 labeled_idx, unlabeled_idx, n_classes, budget,
                 label_penalty: float = -0.05):
        self.images        = train_images
        self.labels        = train_labels
        self.val_loader    = val_loader
        self.n_classes     = n_classes
        self.max_budget    = budget
        self.label_penalty = label_penalty
        # Use list for sampling; convert to set for O(1) removal
        self._init_labeled   = list(labeled_idx)
        self._init_unlabeled = list(unlabeled_idx)
        self.labeled_idx     = list(labeled_idx)
        self.unlabeled_idx   = list(unlabeled_idx)
        self.budget_remaining     = budget
        self.current_sample_idx   = None
        self.current_probs        = None
        self.current_auc          = 0.5
        self.model                = None
        self._query_log           = []

    def reset(self) -> np.ndarray:
        """Reset to initial pools, retrain model from scratch, return first state."""
        self.labeled_idx      = list(self._init_labeled)
        self.unlabeled_idx    = list(self._init_unlabeled)
        self.budget_remaining = self.max_budget
        self._query_log.clear()
        self.model = MedicalCNN(self.n_classes, CONFIG['dropout_rate']).to(DEVICE)
        loader = make_loader(self.images, self.labels, self.labeled_idx, CONFIG['batch_size'])
        train_model(self.model, loader, self.val_loader, self.n_classes)
        self.current_auc = evaluate(self.model, self.val_loader, self.n_classes)['auc']
        return self._next_state()

    def _next_state(self) -> np.ndarray:
        """
        8-dim SOTA state vector.
        Candidate sampling: pick the highest-entropy sample from k=20 candidates
        so the agent always sees the most informative sample available.
        """
        if not self.unlabeled_idx:
            return np.zeros(CONFIG['state_dim'], dtype=np.float32)

        # Sample k candidates and select the most uncertain one
        k          = min(20, len(self.unlabeled_idx))
        candidates = random.sample(self.unlabeled_idx, k)
        imgs_t     = torch.tensor(self.images[candidates], dtype=torch.float32).to(DEVICE)
        with torch.no_grad():
            probs = F.softmax(self.model(imgs_t), dim=-1).cpu().numpy()
        entropies = compute_entropy(probs)
        best_i    = int(np.argmax(entropies))
        self.current_sample_idx = candidates[best_i]
        self.current_probs      = probs[best_i]

        confidence   = float(self.current_probs.max())
        entropy_val  = float(entropies[best_i])
        sorted_p     = np.sort(self.current_probs)[::-1]
        margin       = float(sorted_p[0] - sorted_p[1]) if len(sorted_p) > 1 else 1.0
        budget_ratio = self.budget_remaining / self.max_budget
        auc_norm     = self.current_auc
        labeled_ratio = len(self.labeled_idx) / (len(self.labeled_idx) + len(self.unlabeled_idx) + 1e-6)
        step_ratio    = 1.0 - budget_ratio
        pool_labels   = self.labels[self.labeled_idx]
        counts        = np.bincount(pool_labels, minlength=self.n_classes).astype(float)
        imbalance     = float(counts.min() / (counts.max() + 1e-6))

        return np.array([
            confidence, entropy_val, margin, budget_ratio,
            auc_norm, labeled_ratio, step_ratio, imbalance
        ], dtype=np.float32)

    def step(self, action: int) -> Tuple[np.ndarray, float, bool]:
        """
        Execute action.  
        - Action 1: request annotation (budget cost, shaped reward)
        - Action 0: predict / skip (free, tiny confidence reward)
        """
        prev_auc    = self.current_auc
        reward      = 0.0
        entropy_val = float(compute_entropy(self.current_probs[np.newaxis])[0])                       if self.current_probs is not None else 0.0

        if action == 1 and self.budget_remaining > 0:
            self.unlabeled_idx.remove(self.current_sample_idx)
            self.labeled_idx.append(self.current_sample_idx)
            self.budget_remaining -= 1
            self._query_log.append(1)
            # Entropy proxy reward (immediate, before retraining)
            reward += CONFIG['entropy_proxy_scale'] * entropy_val + self.label_penalty
            # Retrain every 5 labels or at budget exhaustion
            if len(self.labeled_idx) % 5 == 0 or self.budget_remaining == 0:
                loader = make_loader(self.images, self.labels, self.labeled_idx, CONFIG['batch_size'])
                train_model(self.model, loader, self.val_loader, self.n_classes)
                self.current_auc = evaluate(self.model, self.val_loader, self.n_classes)['auc']
                delta_auc = self.current_auc - prev_auc
                reward   += CONFIG['auc_reward_scale'] * delta_auc
        else:
            if self.current_sample_idx in self.unlabeled_idx:
                self.unlabeled_idx.remove(self.current_sample_idx)
            self._query_log.append(0)
            # Tiny reward for skipping a confident sample (budget conservation)
            if self.current_probs is not None:
                reward = 0.01 * float(self.current_probs.max())

        done = (self.budget_remaining <= 0) or (len(self.unlabeled_idx) == 0)
        next_state = self._next_state() if not done else np.zeros(CONFIG['state_dim'], dtype=np.float32)
        return next_state, reward, done

    def get_final_auc(self) -> Tuple[float, float]:
        val_auc  = evaluate(self.model, val_loader,  self.n_classes)['auc']
        test_auc = evaluate(self.model, test_loader, self.n_classes)['auc']
        return val_auc, test_auc


print(" AnnotationEnvironment (SOTA 8-dim state + shaped reward) defined.")
````

## 8. SOTA DQN — LayerNorm Q-Networks + Prioritised Experience Replay


````python
# SOTA Q-Network with LayerNorm

class QNetwork(nn.Module):
    """3-layer MLP with LayerNorm — handles heterogeneous state feature scales."""
    def __init__(self, state_dim: int, action_dim: int, hidden_dim: int = 128):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(state_dim, hidden_dim),
            nn.LayerNorm(hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.LayerNorm(hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, 64),
            nn.ReLU(),
            nn.Linear(64, action_dim),
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.net(x)


# Dueling Q-Network 
class DuelingQNetwork(nn.Module):
    """
    Dueling DQN: Q(s,a) = V(s) + (A(s,a) − mean_a'[A(s,a')])
    Better state-value estimation → faster convergence in sparse-reward tasks.
    """
    def __init__(self, state_dim: int, action_dim: int, hidden_dim: int = 128):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Linear(state_dim, hidden_dim), nn.LayerNorm(hidden_dim), nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim), nn.LayerNorm(hidden_dim), nn.ReLU(),
        )
        self.value_head = nn.Sequential(
            nn.Linear(hidden_dim, 64), nn.ReLU(), nn.Linear(64, 1)
        )
        self.advantage_head = nn.Sequential(
            nn.Linear(hidden_dim, 64), nn.ReLU(), nn.Linear(64, action_dim)
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        feats = self.encoder(x)
        V = self.value_head(feats)
        A = self.advantage_head(feats)
        return V + (A - A.mean(dim=1, keepdim=True))


print(" QNetwork (LayerNorm) and DuelingQNetwork defined.")
print(f"   DuelingQNetwork output: {DuelingQNetwork(8,2)(torch.randn(4,8)).shape}")
````

````python
# Prioritised Experience Replay (PER)
# PER replays high-TD-error transitions more frequently.
# Priority p_i = |δ_i|^α + ε  where δ_i = TD error.
# Importance-sampling (IS) weights correct the bias introduced by non-uniform replay.
# References: Schaul et al. (2015). Prioritized Experience Replay. ICLR 2016.

Transition = collections.namedtuple('Transition',
    ('state', 'action', 'reward', 'next_state', 'done'))


class PrioritisedReplayBuffer:
    """
    Sum-tree-backed PER buffer.
    O(log N) insertion and sampling, O(1) priority update.
    """
    def __init__(self, capacity: int, alpha: float = 0.6):
        self.capacity = capacity
        self.alpha    = alpha
        self.buffer   = []
        self.pos      = 0
        self.priorities = np.zeros(capacity, dtype=np.float32)
        self.max_priority = 1.0

    def push(self, *args):
        if len(self.buffer) < self.capacity:
            self.buffer.append(None)
        self.buffer[self.pos]      = Transition(*args)
        self.priorities[self.pos]  = self.max_priority
        self.pos = (self.pos + 1) % self.capacity

    def sample(self, batch_size: int, beta: float = 0.4):
        n = len(self.buffer)
        probs = self.priorities[:n] ** self.alpha
        probs /= probs.sum()
        indices = np.random.choice(n, batch_size, replace=False, p=probs)
        samples = [self.buffer[i] for i in indices]
        # IS weights to correct sampling bias
        weights = (n * probs[indices]) ** (-beta)
        weights /= weights.max()
        return samples, indices, torch.tensor(weights, dtype=torch.float32).to(DEVICE)

    def update_priorities(self, indices: np.ndarray, td_errors: np.ndarray):
        for idx, err in zip(indices, td_errors):
            p = float(abs(err)) + CONFIG['per_eps']
            self.priorities[idx] = p
            self.max_priority = max(self.max_priority, p)

    def __len__(self):
        return len(self.buffer)


# Keep uniform buffer as fallback for ablations
class UniformReplayBuffer:
    def __init__(self, capacity: int):
        self.buffer = collections.deque(maxlen=capacity)

    def push(self, *args):
        self.buffer.append(Transition(*args))

    def sample(self, batch_size: int, beta: float = None):
        samples = random.sample(self.buffer, batch_size)
        weights = torch.ones(batch_size).to(DEVICE)
        return samples, None, weights

    def update_priorities(self, *args): pass

    def __len__(self): return len(self.buffer)


print("PrioritisedReplayBuffer and UniformReplayBuffer defined.")
````

````python
# Base DQN Agent
class DQNAgent:
    """
    SOTA DQN with:
    - LayerNorm Q-network
    - Soft target updates (τ=0.005, every step)
    - Prioritised Experience Replay
    - Double DQN target computation
    - Gradient clipping
    - IS-weighted loss
    """
    def __init__(self, state_dim: int, action_dim: int, use_per: bool = True):
        self.q_net      = QNetwork(state_dim, action_dim).to(DEVICE)
        self.target_net = QNetwork(state_dim, action_dim).to(DEVICE)
        self.target_net.load_state_dict(self.q_net.state_dict())
        self.target_net.eval()
        self.optimizer  = optim.Adam(self.q_net.parameters(), lr=CONFIG['dqn_lr'])
        self.replay     = (PrioritisedReplayBuffer(CONFIG['replay_buffer_size'], CONFIG['per_alpha'])
                           if use_per else UniformReplayBuffer(CONFIG['replay_buffer_size']))
        self.use_per    = use_per
        self.epsilon    = CONFIG['epsilon_start']
        self.beta       = CONFIG['per_beta_start']
        self.step_count = 0
        self.losses     = []
        self.tau        = CONFIG['tau']

    def select_action(self, state: np.ndarray) -> int:
        if random.random() < self.epsilon:
            return random.randint(0, CONFIG['action_dim'] - 1)
        with torch.no_grad():
            q = self.q_net(torch.tensor(state, dtype=torch.float32).unsqueeze(0).to(DEVICE))
        return int(q.argmax().item())

    def decay_epsilon(self):
        self.epsilon = max(CONFIG['epsilon_end'], self.epsilon * CONFIG['epsilon_decay'])

    def anneal_beta(self, episode: int, total_episodes: int):
        """Linearly anneal IS correction from beta_start to 1.0."""
        frac = episode / max(total_episodes - 1, 1)
        self.beta = CONFIG['per_beta_start'] + frac * (CONFIG['per_beta_end'] - CONFIG['per_beta_start'])

    def update_target(self):
        """Soft update: θ_target ← τ·θ_online + (1−τ)·θ_target."""
        for tp, op in zip(self.target_net.parameters(), self.q_net.parameters()):
            tp.data.copy_(self.tau * op.data + (1 - self.tau) * tp.data)

    def train_step(self) -> Optional[float]:
        if len(self.replay) < CONFIG['dqn_batch_size']:
            return None
        batch, indices, is_weights = self.replay.sample(CONFIG['dqn_batch_size'], self.beta)
        states      = torch.tensor(np.array([t.state      for t in batch]), dtype=torch.float32).to(DEVICE)
        actions     = torch.tensor([t.action    for t in batch], dtype=torch.long).to(DEVICE)
        rewards     = torch.tensor([t.reward    for t in batch], dtype=torch.float32).to(DEVICE)
        next_states = torch.tensor(np.array([t.next_state for t in batch]), dtype=torch.float32).to(DEVICE)
        dones       = torch.tensor([t.done      for t in batch], dtype=torch.float32).to(DEVICE)
        current_q = self.q_net(states).gather(1, actions.unsqueeze(1)).squeeze(1)
        with torch.no_grad():
            # Double DQN: online net selects action, target net evaluates
            best_actions = self.q_net(next_states).argmax(dim=1, keepdim=True)
            next_q       = self.target_net(next_states).gather(1, best_actions).squeeze(1)
            target_q     = rewards + CONFIG['gamma'] * next_q * (1 - dones)
        td_errors = (current_q - target_q).detach().cpu().numpy()
        # IS-weighted Huber loss
        loss = (is_weights * F.smooth_l1_loss(current_q, target_q, reduction='none')).mean()
        self.optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(self.q_net.parameters(), max_norm=10.0)
        self.optimizer.step()
        self.step_count += 1
        self.update_target()   # soft update every step
        if self.use_per and indices is not None:
            self.replay.update_priorities(indices, np.abs(td_errors))
        lv = loss.item()
        self.losses.append(lv)
        return lv


# Double DQN Agent
class DoubleDQNAgent(DQNAgent):
    """Double DQN — already implemented in DQNAgent.train_step(). Subclass exists for naming / ablation purposes."""
    pass


# Dueling DQN Agent 
class DuelingDQNAgent(DQNAgent):
    """Dueling DQN: swaps in DuelingQNetwork, inherits Double DQN + PER training."""
    def __init__(self, state_dim: int, action_dim: int, use_per: bool = True):
        super().__init__(state_dim, action_dim, use_per)
        self.q_net      = DuelingQNetwork(state_dim, action_dim).to(DEVICE)
        self.target_net = DuelingQNetwork(state_dim, action_dim).to(DEVICE)
        self.target_net.load_state_dict(self.q_net.state_dict())
        self.target_net.eval()
        self.optimizer  = optim.Adam(self.q_net.parameters(), lr=CONFIG['dqn_lr'])


print(" DQNAgent (SOTA), DoubleDQNAgent, DuelingDQNAgent defined.")
````
